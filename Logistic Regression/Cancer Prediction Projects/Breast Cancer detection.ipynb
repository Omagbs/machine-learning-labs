## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
RANDOM_STATE = 42
```

## Read Dataset

```python
df = pd.read_csv("data.csv")
```

```python
df.head(20)
```

Note: M = **Malignant**, and B = **Benign**

```python
df.info()
```

```python
df.describe()
```

## Cleaning the Data

```python
sns.heatmap(df.isnull())
plt.show()
```

Unnamed column is empty, so we drop.

```python
df.drop(["Unnamed: 32", "id"], inplace = True, axis = 1)
```

```python
df.head()
```

### Encoding our M as 1 and B as 0, for our diagnosis column

```python
df.diagnosis = [1 if value == "M" else 0 for value in df.diagnosis]
```

```python
df.head()
```

### Changing the data type of the diagnosis column to category

```python
df['diagnosis'] = df['diagnosis'].astype('category')
```

```python
df.info()
```

```python
df['diagnosis'].value_counts().plot(kind='bar')
```

## Assigning predictors and values

```python
y = df['diagnosis']
X = df.drop(['diagnosis'], axis = 1)
```

```python
X
```

## Normalization of Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#fit the scaler to data and transform the data
X_scaled = scaler.fit_transform(X)
```

## Splitting the Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.25, random_state=RANDOM_STATE)
```
